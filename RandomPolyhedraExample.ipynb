```python
# A quick sample construction to showcase how to use koebe.py

from koebe.algorithms.incrementalConvexHull import randomConvexHullE3
from koebe.graphics.spherical2viewer import *

import random

viewer = S2Viewer(600,600)

viewer.toggleSphere() # Hide the sphere
polyhedron = randomConvexHullE3(100)

viewer.addAll(polyhedron.edges)


viewer.show()
```

```python
viewer = S2Viewer(600,600)

viewer.toggleSphere() # Hide the sphere


poly = polyhedron.duplicate()

viewer.addAll(poly.verts)
viewer.addAll(poly.edges)
viewer.addAll(poly.faces)

polyhedron.verts[10].remove()

# Set each face to a random color:
for i in range(len(poly.faces)):
    viewer.setStyle(poly.faces[i], 
                makeStyle(fill = "hsl(" + str(random.randint(0, 255)) 
                                  + ", 100%, 50%)"))

#viewer.setStyle(poly.verts[30],makeStyle(fill = "#f00"))

viewer.show()
```

```python
from koebe.geometries.spherical2 import *
from koebe.graphics.spherical2viewer import *

import random

viewer = S2Viewer(600,600)

disk1 = DiskS2(1000*3,1000*4,1000*1,1000*4.3)
disk2 = DiskS2(-1.05,1,-1,1)
disk3 = disk1.invertThrough(disk2)

viewer.setStyle(disk1, makeStyle(stroke="#ff0000"))
viewer.setStyle(disk3, makeStyle(stroke="#ff0000"))

viewer.add(disk1)
viewer.add(disk2)
viewer.add(disk3)

viewer.show()
```

```python
from koebe.geometries.spherical2 import *
from koebe.geometries.commonOps import *
from koebe.graphics.spherical2viewer import *

import random

viewer = S2Viewer(600,600)

a = DiskS2(1000*3,1000*4,1000*1,1000*4.3)
b = DiskS2(-1.05,1,-1,1)

viewer.setStyle(a, makeStyle(stroke="#f00"))
viewer.setStyle(b, makeStyle(stroke="#00f"))

viewer.add(a)
viewer.add(b)



for _ in [-tick / 10 for tick in range(10)]:
    s = inner_product31(*tuple(a), *tuple(b)) / inner_product31(*tuple(b), *tuple(b))
    c = (a.a - s*b.a, a.b - s*b.b, a.c - s*b.c, a.d - s*b.d)
    cinv = 1.0/math.sqrt(c[0]**2 + c[1]**2 + c[2]**2)
    c = (c[0] * cinv, c[1] * cinv, c[2] * cinv, c[3] * cinv)

    t = 0.005
    b2 = DiskS2(b.a + c[0] * t, b.b + c[1] * t, b.c + c[2] * t, b.d - c[3] * t)
    viewer.setStyle(b2, makeStyle(stroke = "#eee", strokeWeight = 1))
    viewer.add(b2)
    b = b2


viewer.show()
```

```python
from koebe.geometries.spherical2 import *
from koebe.geometries.commonOps import *
from koebe.graphics.spherical2viewer import *

import random

viewer = S2Viewer(1000,1000)

a = DiskS2(1000*3,1000*4,1000*1,1000*4.3)
b = DiskS2(-1.05,1,-1,1)

viewer.setStyle(a, makeStyle(stroke="#f00"))
viewer.setStyle(b, makeStyle(stroke="#00f"))

viewer.add(a)
viewer.add(b)

for _ in [-tick / 10 for tick in range(30)]:
    s = inner_product31(*tuple(a), *tuple(b)) / inner_product31(*tuple(b), *tuple(b))
    c = (a.a - s*b.a, a.b - s*b.b, a.c - s*b.c, a.d - s*b.d)
    cinv = 1.0/math.sqrt(c[0]**2 + c[1]**2 + c[2]**2)
    c = (c[0] * cinv, c[1] * cinv, c[2] * cinv, c[3] * cinv)

    t = -0.1
    b2 = DiskS2(b.a + c[0] * t, b.b + c[1] * t, b.c + c[2] * t, b.d + c[3] * t)
    viewer.setStyle(b2, makeStyle(stroke = "#eee", strokeWeight = 1))
    viewer.add(b2)
    b = b2


viewer.show()
```

```python
from koebe.geometries.spherical2 import *
from koebe.graphics.spherical2viewer import *

import random

viewer = S2Viewer(600,600)

d1 = 0.75
d2 = -0.5

disk1 = DiskS2(0.0, 0.0, 1.0, d1)

z2 = -math.sqrt(1 - d1 * d1) * math.sqrt(1 - d2 * d2) + d1 * d2
x2 = math.sqrt(1 - z2 * z2)
disk2 = DiskS2(x2, 0.0, z2, d2)

viewer.add(disk1)
viewer.add(disk2)

viewer.setStyle(disk1, makeStyle(stroke="#f00"))
viewer.setStyle(disk2, makeStyle(stroke="#00f"))

viewer.show()
```
