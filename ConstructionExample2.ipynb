```python
from koebe.geometries.spherical2 import *
from koebe.graphics.spherical2viewer import *

viewer = S2Viewer(600,600)

disk0 = DiskS2(1.05,0,0,1)
disk1 = DiskS2(0,1.05,0,1)
disk2 = DiskS2(0,0,1.05,1)

viewer.addAll([disk0, disk1, disk2])
viewer.addAll([d.directionE3.v.toPointE3() for d in [disk0, disk1, disk2]])
for d in [disk0, disk1, disk2]:
    print(tuple(d.normedBasis3.v))
for v in [d.directionE3.v for d in [disk0, disk1, disk2]]:
    print(tuple(v))
viewer.show()
```
